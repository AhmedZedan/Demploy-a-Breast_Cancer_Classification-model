```python
df = pd.read_csv('breast_cancer.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df['target'] = df['target'].astype(int)
```

```python
df['target'].value_counts()
```

```python
sns.countplot(x=df["target"])
```

```python
sns.pairplot(df, hue = 'target', vars = ['mean radius', 'mean texture', 'mean area', 'mean perimeter', 'mean smoothness'])
```

```python
plt.figure(figsize=(20, 15))
sns.heatmap(df.corr(), annot = True)
```

```python
y = df['target']
```

```python
X = df.drop('target', axis=1)
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
svc_model = SVC()
```

```python
svc_model.fit(X_train, y_train)
```

```python
y_pred = svc_model.predict(X_test)
```

```python
y_pred
```

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, y_pred))
```

```python
print(accuracy_score(y_test, y_pred))
```

## Now we will do some preprocessing on the data:
We will scaling the data to make it in range 0->1 and see how this will affect the model

```python
from sklearn import preprocessing
```

```python
scaler = preprocessing.MinMaxScaler()
scaler_obj = scaler.fit(X_train)
```

```python
X_train_scaled = scaler_obj.transform(X_train)
```

```python
X_train_scaled
```

```python
X_test_scaled = scaler_obj.transform(X_test)
```

```python
X_test_scaled
```

```python
scaler_filename = "scaler.pkl"
joblib.dump(scaler_obj, scaler_filename)
```

```python
svc_model.fit(X_train_scaled, y_train)
```

```python
y_pred_scale = svc_model.predict(X_test_scaled)
```

```python
y_pred_scale
```

```python
cm_scale = confusion_matrix(y_test, y_pred_scale)
```

```python
ax = plt.axes()
sns.heatmap(cm_scale,annot=True, ax = ax)
ax.set_title('HeatMap of confusion_matrix after scaling')
plt.show()
```

```python
print(classification_report(y_test, y_pred_scale))
```

```python
print(accuracy_score(y_test, y_pred_scale))
```

## Now we will our new model with accuracy 98%

```python
model_name = "model.pkl"
joblib.dump(svc_model, model_name)
```
